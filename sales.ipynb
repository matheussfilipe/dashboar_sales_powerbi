```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_sales_analysis.csv')

df
```

```python
## info do dataset
df.info()
```

```python
## Adicionando uma nova coluna com o total de vendas

df["total_sales"] = df.loc[:,"sales_month_1":"sales_month_12"].sum(axis=1)

df.head()
```

```python
## Verificando a existência de dados nulos
df.isnull().sum()
```

```python
## Verificando a existência de dados duplicados
df.duplicated().sum()
```

```python
## Adicionando uma coluna da receita 

df['total_revenue'] = df['price']*df['total_sales']

df.head()
```

```python
## Ordenando a base de dados em relação ao número de vendas 
df.sort_values('total_sales', ascending = False, inplace = True)

df.reset_index(drop = True, inplace = True)

df
```

```python
## df para o power BI

df.to_csv('sales.csv')
```
